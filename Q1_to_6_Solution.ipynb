```python
#Euler Method for Malthusian Population Growth Model
#dN(t)/dt = r*N(t), N(t): population at time t, r: Growth rate 
```

```python
#import the required packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(N_zero, growth_rate, h, final_T):
    N_t = N_zero
    euler_iters = [] #iterates obtained using Euler Method
    
    num_intervals = final_T/h  # it gives the number of iterations or n=(b-a)/h => b = a + n*h
    for i in range(int(num_intervals)):  # for h = 0.00001, final_T/h = 1.0/0.00001 = 99999.99999999999 != 100000. => int(num_intervals) = 99999 
        euler_iters.append(N_t)
        N_t = N_t + h *  growth_rate * N_t  # predicted population

    # this modification removes the error.
    if h==0.00001:            
      euler_iters.append(N_t)    
    print('euler predicted at t=1: ', (N_t))
    euler_predicted.append(N_t)

    print('actual value at t=1: ', (N_zero * np.exp(growth_rate * (final_T))))
    actual_value.append(N_zero * np.exp(growth_rate * (final_T)))

    print("Error: ",N_zero * np.exp(growth_rate * (final_T)) - N_t)
    error_value.append(N_zero * np.exp(growth_rate * (final_T)) - N_t)
    
    #x and y axes to plot exp curve
    xx = np.arange(0,final_T,h)  #creates an array of t values with increment h
    yy = N_zero*np.exp(growth_rate*xx) # Actual population at time t
    plt.plot(xx,yy,color='red')
    
    #plot the Euler method's predicted values
    plt.plot(xx,euler_iters,color='blue')
    
    #other plot related details
    plt.xlabel('Time')
    plt.ylabel('N_t')
    plt.legend(['Actual','Euler'])
    plt.show()
        
    return 1
```

${\Large\textbf{2, 3, & 4.QUESTION}}$

```python
N_zero = 100.0
growth_rate = 0.9
euler_h = [0.1, 0.01, 0.001, 0.0001, 0.00001]
final_T = 1.0

euler_predicted = []
actual_value = []
error_value = []

for h in euler_h:
  euler(N_zero, growth_rate, h, final_T)
```

**COMMENT ON THE OBSERVATIONS** \
1. From the above table it is clear that as we go on decreasing the value of h we tend to a better or an improved approximation. Consequently error value decreases. So, for higher values of h, the predicted value deviates from the actual value which is quite obvious from the graph for h=0.1
2. Also empirically (In fact from the above obtained error data) it is clear that the errors behave as  $\sim O(h^2)$.
